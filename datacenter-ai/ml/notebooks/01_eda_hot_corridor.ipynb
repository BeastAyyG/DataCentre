# 01 — EDA: Data Centre Hot Corridor Temperature

Exploratory analysis of server inlet/outlet temperatures, power, and airflow.


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style('whitegrid')
import warnings; warnings.filterwarnings('ignore')

df = pd.read_csv('../data/raw/hot_corridor.csv')
df['timestamp'] = pd.to_datetime(df['timestamp'])
df = df.sort_values('timestamp')
print(f'Shape: {df.shape}')
df.describe()
```

```python
# Temperature distributions
fig, axes = plt.subplots(1, 3, figsize=(15, 4))
for ax, col in zip(axes, ['inlet_temp_c', 'outlet_temp_c', 'power_kw']):
    if col in df.columns:
        df[col].hist(bins=50, ax=ax, alpha=0.7, color='steelblue')
        ax.set_title(col)
plt.tight_layout()
plt.savefig('../artifacts/01_temp_dist.png', dpi=150)
plt.show()
```

## Key Findings
- Inlet temps typically 18–28°C with business-hours peak
- Power shows clear diurnal pattern
- Outlet temp consistently 2–4°C above inlet
