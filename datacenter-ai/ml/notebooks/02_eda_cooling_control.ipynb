# 02 — EDA: Data Center Cooling Control Dataset

CRAC unit performance, chiller efficiency, and setpoint dynamics.


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style('whitegrid')
import warnings; warnings.filterwarnings('ignore')

df = pd.read_csv('../data/raw/cooling_control.csv')
df['timestamp'] = pd.to_datetime(df['timestamp'])
df = df.sort_values('timestamp')
print(f'Shape: {df.shape}')
df.describe()
```

```python
# Correlation heatmap
cols = ['inlet_temp_c','outlet_temp_c','power_kw','cooling_output_kw','airflow_cfm','humidity_pct']
avail = [c for c in cols if c in df.columns]
plt.figure(figsize=(8,6))
sns.heatmap(df[avail].corr(), annot=True, fmt='.2f', cmap='coolwarm', center=0)
plt.title('Cooling System — Feature Correlations')
plt.tight_layout()
plt.savefig('../artifacts/02_correlation.png', dpi=150)
plt.show()
```

## Key Findings
- Cooling output tightly coupled with power consumption
- Temperature delta (outlet–inlet) is a strong cooling inefficiency proxy
- Prophet's daily seasonality will capture time-of-day patterns well
